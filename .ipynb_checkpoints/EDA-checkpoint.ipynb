## Analisi di una variabile categorica (Contract)

```python
df['Contract'].value_counts().plot(kind='bar')
plt.title("Tipi di contratto")
plt.xlabel("Tipo di contratto")
plt.ylabel("Numero di clienti")
plt.show()
```

---

## Churn rate per tipo di contratto

Mostra quali contratti hanno il churn più alto

```python
(
    df.groupby('Contract')['Churn']
    .apply(lambda x: (x=='Yes').mean())
    .sort_values(ascending=False)
    .plot(kind='bar')
    
)

plt.title("Churn rate per tipo di contratto")
plt.ylabel("Percentuale di clienti che abbandonano")
plt.show()
```
